## **4. Data Validation - Great Expectations**

```python
# Initialize data context object

context = gx.get_context()
```

```python
# Add data source according to method of loading (Pandas)

datasource_name = 'supply_chain_data-csv'
datasource = context.sources.add_pandas(datasource_name)
```

```python
# Create asset name based on the data loading method and file path provided

asset_name = 'supply-chain-data'
path = './P2M3_frans_joddy_sutjianto_data_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path)
```

```python
# Create a batch request based on the asset loaded

batch_req = asset.build_batch_request()
```

```python
# Create an expectation suite to store all validation used

expectation_suite_name = 'expectation-supply-chain-data'
context.add_or_update_expectation_suite(expectation_suite_name)

validator = context.get_validator(batch_request=batch_req, expectation_suite_name=expectation_suite_name)
```

```python
# Check if revenue generated column values are unique

validator.expect_column_values_to_be_unique('revenue_generated')
```

```python
# Check if price column's value is between 0 and 100

validator.expect_column_values_to_be_between(column='price', min_value=0, max_value=100)
```

```python
# Check if inspection_results distinct value is in set

validator.expect_column_distinct_values_to_be_in_set(column='inspection_results', value_set=['Pending', 'Fail', 'Pass'])
```

```python
# Check if defect_rates column data type is float

validator.expect_column_values_to_be_in_type_list(column='defect_rates', type_list=['float'])
```

```python
# Check if routes column value length is equal to 7

validator.expect_column_value_lengths_to_equal(column='routes', value=7)
```

```python
# Check if column costs' mean does not exceed column revenue_generated's mean

validator.expect_column_mean_to_be_between(column='costs', min_value=0, max_value=5800)
```

```python
# Check if customer_demographics column's proportion unique value falls between 0 and 0.4

validator.expect_column_proportion_of_unique_values_to_be_between(column='customer_demographics', min_value=0, max_value=0.4)
```
